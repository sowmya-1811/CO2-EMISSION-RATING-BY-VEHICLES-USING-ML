```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned
```

```python
df_cleaned.info()
```

```python
# Verify that duplicates have been removed
print("Shape of the original DataFrame:", df.shape)
print("Shape of the DataFrame after removing duplicates:", df_cleaned.shape)
```

```python
# Get unique values for categorical columns (object type)
for col in df_cleaned.select_dtypes(include='object').columns:
    print(df_cleaned[col].value_counts)
    print("------"*10)
```

```python
for col in df_cleaned.select_dtypes(exclude='object').columns:
    print(df_cleaned[col])
    print("------"*10)
```

```python
# Get descriptive statistics for numerical columns to identify outliers
df_cleaned.describe().T
```

```python
df_cleaned.describe(include="object")
```

```python
df_cleaned.to_csv("CO2_Emissions_Cleaned.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='Engine Size(L)', y='CO2 Emissions(g/km)', palette='husl', hue='Cylinders')
plt.title('Scatter Plot of Engine Size vs. CO2 Emissions by Cylinders')
plt.xlabel('Engine Size(L)')
plt.ylabel('CO2 Emissions(g/km)')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Fuel Type', y='CO2 Emissions(g/km)', data=df_cleaned, palette='Set2')
plt.title('CO₂ Emissions by Fuel Type', fontsize=14)
plt.xlabel('Fuel Type')
plt.ylabel('CO₂ Emissions (g/km)')
plt.show()
```

```python
corr = df_cleaned[['Engine Size(L)', 'Cylinders', 'Fuel Consumption Comb (L/100 km)', 'CO2 Emissions(g/km)']].corr()
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```
